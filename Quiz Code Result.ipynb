```python
import pandas as pd
url='http://bit.ly/HDSC-Dataset'
fuel_data=pd.read_csv(url, error_bad_lines=False)
fuel_data.describe(include='all')
```

```python
fuel_data
```

```python
# to find the lowest average cost per unit burned
# you first group it by using a groupby() method
fuel_data.groupby('fuel_cost_per_unit_burned')['fuel_cost_per_unit_burned'].count()
fuel_data[['fuel_cost_per_unit_burned']]=fuel_data[['fuel_cost_per_unit_burned']].min().mean()
print (fuel_data[['fuel_cost_per_unit_burned']])
```

```python
# question 4: standard deviation and the 75th percentile of(fuel_mmbtu_per_unit)
fuel_data.groupby('fuel_mmbtu_per_unit')['fuel_mmbtu_per_unit'].count()
fuel_data[['fuel_mmbtu_per_unit']].std()
```

```python
fuel_data.groupby('fuel_mmbtu_per_unit')['fuel_mmbtu_per_unit'].count()
fuel_data[['fuel_mmbtu_per_unit']].describe()
```

```python
#questio 5 for skewness and kurtosis
fuel_data.groupby('fuel_qty_burned')['fuel_qty_burned'].count()
fuel_data[['fuel_qty_burned']].skew()
```

```python
fuel_data.groupby('fuel_qty_burned')['fuel_qty_burned'].count()
fuel_data[['fuel_qty_burned']].kurt()
```
